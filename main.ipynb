```python
# Filter positive weights for pie chart
positive_systems = [s for s, w in zip(systems, weights) if w > 0]
positive_weights = [w for w in weights if w > 0]

# Calculate percentages
total_positive = sum(positive_weights)
percentages = [(w/total_positive)*100 for w in positive_weights]

# Create figure with tight layout
fig, ax = plt.subplots(figsize=(10, 8))

# Define a nice color palette (pastel colors similar to the example)
colors = ['#f35163', '#e87a55', '#fdae61', '#fee08b', '#ffffbf', 
          '#e6f598', '#abdda4', '#66c2a5', '#3288bd']

# Create pie chart without labels and percentages, with white edges
wedges, texts = ax.pie(
    positive_weights, 
    colors=colors[:len(positive_systems)],
    startangle=90,
    wedgeprops={'edgecolor': 'white', 'linewidth': 2}
)

# Add legend with colored squares
ax.legend(wedges, positive_systems,
        #   title="Components",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1),
          fontsize=12,
          title_fontsize=14,
          frameon=False)

# Remove extra whitespace
ax.axis('equal')
# plt.show()
plt.savefig('res/figures/table2.png', dpi=500, bbox_inches='tight', pad_inches=0.05, transparent=False)
```
